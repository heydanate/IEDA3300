## In this exercise, please first load the alchol.csv dataset as a dataframe and then finish the following questions

```python
data4 = pd.read_csv('alcohol.csv')
data4
```

Q1. Clean the dataset by removing all rows that contain missing values. Use the cleaned dataset to do subsequent analysis.

```python
data4.dropna(inplace = True)
data4
```

Q2. Calculate the mean, average, and standard deviation of alcohol consumption over the year based on WHO region. Use a datafram to present the result.

```python
table = data4.groupby('WHO region').agg({'Display Value':['mean','std']})
#groupby group by column then agg solves mean median std
table
```

Q3. Select the country with the largest and the smallest averaged consumption (over the year) on beer

```python
#filter row where beverage type beer
filterbeer = data4[data4['Beverage Types'] == 'Beer']

#group the filter row by WHO and average display value of each group
group = filterbeer.groupby('Country')['Display Value'].mean()
maxbeer = group.idxmax()
minbeer = group.idxmin()

print('The largest averaged consumption (over the year) on beer is:', maxbeer)
print('The smallest averaged consumption (over the year) on beer is:', minbeer)
```

Q4. For each country, calculate the difference between total alcohol consumption (all types) before (including) 1986 and after 1986. When there is no data (e.g. no data of China before 1986), treat it as zero. Use a dataframe of two columns to present the result 

```python
bef1986 = data4[data4['Year'] <= 1986]
aft1986 = data4[data4['Year'] > 1986]

# Calculate the difference in alcohol consumption for each country
diff = []
for country in data4['Country'].unique():
    before = bef1986[bef1986['Country'] == country]['Display Value'].fillna(0).sum()
    after = aft1986[aft1986['Country'] == country]['Display Value'].fillna(0).sum()
    differ = after - before
    diff.append({'Country': country, 'Difference': differ})

# Create a dataframe to store the results
data4q = pd.concat([pd.DataFrame(diff)], ignore_index = True)

data4q
```

Q5. For each country and each year, define a variable 'trend' = 1 if corresponding alcohol consumption (all types) is larger than the previous year. For example, when the value of China in 1986 is larger than 1985, then the 'trend' variable of China in 1986 is 1. Otherwise, set 'trend'=-1. Same as Q4, when there is no data, treat is as zero. Calculate the 'trend' variable and add it as a column to the original dataframe

```python
data4.sort_values(['Country', 'Year'], inplace = True)
data4['trend'] = -1

for (region, year), group in data4.groupby(['Country','Year']):
    prev_value = group['Display Value'].shift(1)
    year_trend = (group['Display Value'] > prev_value).replace({True: 1, False: -1}).fillna(0)
    data4.loc[group.index, 'trend'] = year_trend
    
data4
```

Q6. In each year, find the WHO-region that contains the most countries whose 'trend'==1. Present the result as a dataframe 

```python
data4q6 = pd.DataFrame(columns = ['Year', 'WHO region', 'Count'])

for year, group in data4.groupby('Year'):
    trendcount = group[group['trend'] ==1].groupby('WHO region')['Country'].nunique()
    maxcount = trendcount.max()
    maxregion = trendcount[trendcount == maxcount].index.tolist()
    data4q6 = pd.concat([data4q6, pd.DataFrame({'Year':[year], 'WHO region':[','.join(maxregion)], 'Count':[maxcount]})], ignore_index = True)
    
data4q6
```
